```python
import random

while True:
  # welcome message
  print('=================================')
  print('Welcome to the number guessing game!')
  print('I am thinking of a number between 1 and 100.')
  print('You have 10 attempts. Good luck! ')

  secret = random.randint(1, 100)
  attempts = 0

  while True:

    try:
        guess = int(input('Enter your guess: '))
    except ValueError:
        print('Please enter a number only!')
        continue

    attempts += 1

    remaining = 10 - attempts

    if guess == secret:
     print(f'You guessed the secret number {secret} in {attempts} attempts!')
     break

    elif guess < secret:
      print(f'The secret number is higher than {guess}')


    elif guess > secret:
      print(f'The secret number is lower than {guess}')

    if attempts == 10:
      print(f'Game over! The number was {secret}')
      break

play_again = input('Play again? yes/no: ')
if play_again == 'yes':
  print(f'Starting new game! ')
else:
   print('Thanks for playing. Goodbye! ')




```
